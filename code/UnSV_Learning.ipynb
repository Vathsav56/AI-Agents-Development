```python
import pandas as pd
```

```python
try:
  dt = pd.read_csv('/content/Mall_Customers.csv')
  print("File Loaded Successfully")
except:
  print("File loading error :-- Either the path is Wrong are the file isnt loaded into the runtime")
```

```python
dt.info()
```

```python
x = dt["Annual Income (k$)"]
y = dt["Spending Score (1-100)"]
```

```python
 x = dt.iloc[:,3:]
 x.head()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5,init='k-means++',random_state=42)
kmeans.fit(x)
```

```python
lables = kmeans.labels_
dt['cluster'] = lables
```

```python
plt.scatter(dt['Annual Income (k$)'], dt['Spending Score (1-100)'], c= dt['cluster'])
plt.title("Mall customer Segmentation")

plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.show()
```
